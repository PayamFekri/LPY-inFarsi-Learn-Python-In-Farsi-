```python
afg.assign(col_divide = lambda x : (x["birth rate"]/ x["death rate"])) #تفاوتش با بالایی اینه که این سطر به سطر تقسیم میکنه ولی جواب یکی هست
```

query over the dataset

```python
afg.query("GDP>15856600000")
```

```python
afg.query("GDP>15856600000").assign(
    GDP_tarakomJamiiat = lambda x : (x["GDP"]/x["tarakom jamiat"])
        )
```

```python
(afg.query("GDP>15856600000").assign(
    GDP_tarakomJamiiat = lambda x : (x["GDP"]/x["tarakom jamiat"]),
    useInternet_unempolyment= lambda x :(x["use internet"]/x["Unemployment"])
        )
    )
```

```python
afg.query("GDP>15856600000").assign(
    GDP_tarakomJamiiat = lambda x : (x["GDP"]/x["tarakom jamiat"]),
    useInternet_unempolyment= lambda x :(x["use internet"]/x["Unemployment"])
        ).plot(kind="scatter" ,x = "GDP_tarakomJamiiat" , y = "useInternet_unempolyment")
```
