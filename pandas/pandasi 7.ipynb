```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'key' :['k0','k1','k2','k3','k4','k5'],
    'A' :  ['A0','A1','A2','A3','A4','A5']
})
df1
```

```python
df2 = pd.DataFrame({
    'key' :['k0','k1','k2'],
    'B' :  ['B0','B1','B2']
})
df2
```

```python
df1.join(df2,lsuffix='_caller' , rsuffix='_other')
```

```python
df1.join(df2.set_index('key'),on='key')
```

# Merge function
<font face = 'XB Zar' color = 'black' size = 3>دقیقا ستون چپ و راست را که روی ان عملیات می کنیم می توان در آن انتخاب کرد 
 </font> 


```python
dfm1 = pd.DataFrame({
    'L-key' :['A','B','C','F'],
    'Value' :  [1,2,3,4]
})
dfm1
```

```python
dfm2 = pd.DataFrame({
    'R-key' :['A','B','C','D'],
    'Value' :  [4,5,6,7]
})
dfm2
```

<b>inner join</b> - default:

```python
dfm1.merge(dfm2,left_on='L-key' , right_on='R-key',how='inner')
```

<b>Left join</b>:

```python
dfm1.merge(dfm2,left_on='L-key' , right_on='R-key',how='left')
```

<b>Right join</b>:

```python
dfm1.merge(dfm2,left_on='L-key' , right_on='R-key',how='right')
```

# <font color = 'green' size = 4><b>concat function - چسباندن(الحاق)</b></font> 

<font face = 'Verdana' color = 'black' size = 2> concatination : in concat we do NOT have left or right join
;just have  inner or outer(default) join </font>

```python
dfc1 = pd.DataFrame({
    'key' :['B','B','A','C'],
    'Value1' :  [1,2,3,4]
})
dfc1
```

```python
dfc2 = pd.DataFrame({
    'key' :['A','B','D'],
    'Value2' :  [1,2,3]
})
dfc2
```

```python
pd.concat([dfc1,dfc2])
```
