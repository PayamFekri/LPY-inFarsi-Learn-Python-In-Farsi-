```python
Aq = pd.read_csv("Air_Quality.csv")
Aq
```

```python
no2_sub = Aq[Aq['Name'] == "Nitrogen dioxide (NO2)"]
no2_sub
```

```python
no2_place = no2_sub.sort_index().groupby(["Geo Place Name"]).head(2)
# ÛŒØ¹Ù†ÛŒ Ø§Ø² Ù‡Ø± Ú¯Ø±ÙˆÙ‡ 2 ØªØ§ Ù…Ù‚Ø¯Ø§Ø± Ø§ÙˆÙ„Ø´Ùˆ Ú©Ù‡ Ú¯Ø±ÙˆÙ‡ Ø¨Ù†Ø¯ÛŒ Ú©Ø±Ø¯ÛŒ Ø¨Ù‡ Ù…Ù† Ø¨Ø¯Ù‡
no2_place
```

<h1 font = 'arial'>pivot , pivot table , melt for reshaping</h1>

```python
Name_dataValue=no2_sub.sort_index().pivot(columns= "Name",
                                           values='Data Value' ).plot()
Name_dataValue
```

```python
place_dataValue=no2_sub.sort_index().pivot(columns= "Geo Place Name",
                                           values='Data Value' ).plot()
place_dataValue
```

```python
aq0 = Aq.pivot_table(
    values= 'Data Value' ,
    index= "Geo Place Name" ,
    columns= 'Name',
    aggfunc='mean' ,
    margins= True
)
aq0
```
