#### Decode & Encode Data

```python
Code = {
    'a': '!',
    'b': 'v',
    'c': '6',
    'd': '(',
    'e': 'w',
    'f': '1',
    'g': ']',
    'h': 'A',
    'i': '=',
    'j': '?',
    'k': 'N',
    'l': '[',
    'm': '@',
    'n': '-',
    'o': '/',
    'p': '#',
    'q': '2',
    'r': '%',
    's': '*',
    't': '+',
    'u': '3',
    'v': 'r',
    'w': 'y',
    'x': '4',
    'y': '&',
    'z': ')',
    '.': 'U ',
    ' ': '_',
    'A': '2',   
}
```

```python
#Encoding
x = "payam fekri"
y = ""
for i in x:
    for j in Code:
        if i==j :
            y = y + Code[i]
y
```

```python
#Decoding
z = ""
for i in y:
    for j in Code:
        if i== Code[j]:
            z = z + j
z
```

#### Decode & Encode specefic word in text

```python
code = {
    "python" : "--1/",
    "AI" : "f7>"
}
```

```python
x = 'python is good for build AI agent'

for i in x.split():
    for j in code:
        if i==j:
            x = x.replace(i,code[i])
x
```

```python
Y = x

for i in Y.split():
    for j in code:
        if i == code[j]:
            Y = Y.replace(i, j)
            
Y
```
