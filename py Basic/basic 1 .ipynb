```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_gate(ax, gate_type, center):
    x, y = center
    if gate_type == "AND":
        ax.add_patch(patches.Rectangle((x-0.5, y-0.3), 0.5, 0.7, fill=True, lw=2, facecolor='black'))
        ax.add_patch(patches.Arc((x, y+0.04), 0.7, 0.7, theta1=-90, theta2=90, lw=2, color='black'))
        ax.text(x-0.2, y-0.7, "AND", fontsize=8, ha="center", color='white')  # متن سفید روی مشکی
    
    elif gate_type == "OR":
        ax.add_patch(patches.Arc((x-0.4, y), 1.0, 1.0, theta1=-65, theta2=65, lw=2, color='black'))
        ax.add_patch(patches.Arc((x-0.7, y), 1.4, 1.4, theta1=-45, theta2=45, lw=2, color='black'))
        ax.text(x-0.2, y-0.7, "OR", fontsize=8, ha="center", color='white')
    
    elif gate_type == "NOT":
        ax.add_patch(patches.Polygon([[x-0.5, y-0.3], [x-0.5, y+0.3], [x, y]], closed=True, fill=True, lw=2, facecolor='black'))
        ax.add_patch(patches.Circle((x+0.1, y), 0.08, fill=True, lw=2, facecolor='black'))
        ax.text(x-0.2, y-0.7, "NOT", fontsize=8, ha="center", color='white')

def draw_interactive_circuit(A_val, B_val, C_val):
    fig, ax = plt.subplots(figsize=(9, 5))
    ax.set_xlim(-4, 6)
    ax.set_ylim(-2, 2)
    ax.axis("off")

    and_out = A_val and B_val
    not_out = not C_val
    or_out = and_out or not_out

    def line_color(val):
        return "green" if val else "red"

    ax.text(-3.8, 1, f"A={A_val}", fontsize=12)
    ax.text(-3.8, 0, f"B={B_val}", fontsize=12)
    ax.text(-3.8, -1, f"C={C_val}", fontsize=12)
    
    # خطوط ورودی
    ax.plot([-3.3, -2], [1, 1], color=line_color(A_val), lw=2)
    ax.plot([-3.3, -2], [0, 0], color=line_color(B_val), lw=2)
    ax.plot([-3.3, -2], [-1, -1], color=line_color(C_val), lw=2)

    # گیت AND
    draw_gate(ax, "AND", (-1.5, 0.5))
    ax.plot([-2, -1.5], [1, 0.65], color=line_color(A_val), lw=2)
    ax.plot([-2, -1.5], [0, 0.35], color=line_color(B_val), lw=2)

    # گیت NOT
    draw_gate(ax, "NOT", (-1.5, -1))
    ax.plot([-2, -1.5], [-1, -1], color=line_color(C_val), lw=2)

    # گیت OR
    draw_gate(ax, "OR", (2, -0.25))
    ax.plot([-1.2, 1.6], [0.5, -0.05], color=line_color(and_out), lw=2)
    ax.plot([-1.2, 1.6], [-1, -0.45], color=line_color(not_out), lw=2)

    # خروجی
    ax.plot([2.4, 4], [-0.25, -0.25], color=line_color(or_out), lw=2)
    ax.text(4.1, -0.25, f"Y={int(or_out)}", fontsize=12)

    plt.title("Interactive Logic Circuit: Y = (A AND B) OR (NOT C)")
    plt.tight_layout()
    plt.show()

A_val = 1 #int(input("Enter A (0 or 1): "))
B_val = 1 #int(input("Enter B (0 or 1): "))
C_val = 1  #int(input("Enter C (0 or 1): "))

draw_interactive_circuit(A_val, B_val, C_val)
```

```python
# استخراج موجودیت های نامدار از متن
List_Name_Person = {'فرهاد', 'جواد', 'رضا', 'عباس', 'رسول'}
List_Family_Person = {'سعیدی', 'جواد', 'ابراهیمی', 'محمدی', 'احمدی'}
List_City = {'ایلام', 'ارومیه', 'سنندج','مازندران', 'مشهد', 'تهران', 'کرمانشاه'}
List_Univercity = {'کردستان','رازی', 'آزاد', 'شریف', 'تهران'}
List_Expertise = {'AI', "NLP", '', 'C++', 'python', 'Java', 'Django', 'Flask', 'MATLAB'}

name_person = set()
family_person = set()
city = set()
univercity = set()
expertise = set()


email = '''
سلام وقتتون بخیر بابت آگهی مزاحم تون شدم اسم بنده رضا ابراهیمی است،
فارغ التحصیل مهندسی کامپیوتر از دانشگاه تهران. محل سکونت هم در شمال تهران 
و دارای تخصص های متفاوتی هستم مانند: NLP , C++ , Django .
    ممنون میشم روزمه بنده رو مطالعه و به من اطلاع دهید. خیلی ممنون.

'''

for i in email.split():
    
    for np in List_Name_Person:
        if i == np:
            name_person.add(np)
            
    for fp in List_Family_Person:
        if i == fp:
            family_person.add(fp)
            
    for c in List_City:
        if i == c:
            city.add(c)
            
    for u in List_Univercity:
        if i == u:
            univercity.add(u)
            
    for e in List_Expertise:
        if i == e:
            expertise.add(e)
```
